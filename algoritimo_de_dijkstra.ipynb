# Imports

```python
import networkx as nx
import matplotlib.pyplot as plt
```

# Criação do grafo com networkxx

```python
# Criação do grafo direcionado
G = nx.DiGraph()

# Adicionar os vértices e arestas com custo
edges = [
    ('A', 'B', 2),
    ('A', 'C', 4),
    ('A', 'D', 7),
    ('B', 'E', 6),
    ('C', 'F', 3),
    ('D', 'E', 5),
    ('C', 'E', 1),
    ('E', 'F', 4)
]

# Adiciona as arestas com os custos
for u, v, cost in edges:
    G.add_edge(u, v, weight=cost)

# Desenhar o grafo
pos = nx.spring_layout(G)  # posição dos vértices
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', font_size=10, font_weight='bold', arrowsize=20)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# Mostrar o grafo
plt.title("Grafo Direcionado com Custos")
plt.show()
```

# Analise da distância

```python
no_inicial = 'A'
distancia_para_cada_no = {key: (float('inf') if key != no_inicial else 0) for key in list(G.nodes)}
nos_nao_visitados = list(G.nodes)

while nos_nao_visitados != []:
  distancia_nos_nao_visitados = {key: distancia_para_cada_no[key] for key in nos_nao_visitados}
  no_nao_visitado_com_menor_valor = min(distancia_nos_nao_visitados, key=distancia_nos_nao_visitados.get)
  no_atual = no_nao_visitado_com_menor_valor

  for vizinho in list(G.successors(no_atual)):
    distancia = distancia_para_cada_no[no_atual] + G[no_atual][vizinho]['weight']
    if distancia < distancia_para_cada_no[vizinho]:
      distancia_para_cada_no[vizinho] = distancia

  nos_nao_visitados.remove(no_nao_visitado_com_menor_valor)

distancia_para_cada_no
```

```python
pos = nx.spring_layout(G)  # posição dos vértices
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', font_size=10, font_weight='bold', arrowsize=20)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.show()
```
